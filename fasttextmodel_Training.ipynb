```python
import fasttext
```

```python
import random
random.seed(42)
```

```python
model1 = fasttext.train_supervised(input='waf.train')

# evaluate the model
result1 = model1.test('waf.test')

# print the evaluation metrics
print(result1)
```

We can provide an autotune validation file which is used during the model training process to automatically select the best hyperparameters for the model.

```python
model = fasttext.train_supervised(input='waf.train', autotuneValidationFile='waf.val')
```

```python
result = model.test('waf.test')
print(result)
```

We obtain way better results

```python
model.save_model("fasttextmodel.bin")
```
